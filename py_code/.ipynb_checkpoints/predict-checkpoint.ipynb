```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.svm import SVR
from sklearn import ensemble
from sklearn import preprocessing
import csv
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
# loading the data
FNAME = '../data/new_york_airbnb/ny_first1000.csv'
data = []
with open(FNAME) as f:
    reader = csv.DictReader(f)
    for row in reader:
        data.append(dict(row))
```

```python
data
```

```python
# makes x and y datasets
def make_x_y(data):
    # extracting all the labels
    borough_labels = list(set([i['borough'] for i in data]))
    neighborhood_labels = list(set([i['neighborhood'] for i in data]))

    borough_le = preprocessing.LabelEncoder()
    borough_le.fit(borough_labels)    
    
    neighborhood_le = preprocessing.LabelEncoder()
    neighborhood_le.fit(neighborhood_labels)    
    
    X = [ [ 
           float(i['latitude']),
           float(i['longitude']),
           float(i['accommodates']),
           borough_le.transform([ i['borough'] ]),
           neighborhood_le.transform([ i['neighborhood'] ]),
        
          ] 
         for i in data]
    y = [float(i['price']) for i in data]
    return (np.array(X), np.array(y))
```

```python
X, y = make_x_y(data)
```

```python
# split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=99)
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
X_train
```

```python
# fit the model
#regr = svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
regr = ensemble.RandomForestRegressor()
regr.fit(X_train, y_train)
```

```python
preds_train = regr.predict(X_train)
preds = regr.predict(X_test)
mse_train = mean_squared_error(y_train, preds_train)
mse_test = mean_squared_error(y_test, preds)

print('MSE train: ', mse_train)
print('MSE test: ', mse_test)
```
